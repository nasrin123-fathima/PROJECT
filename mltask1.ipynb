```python
x=house_price[['bedrooms',	'bathrooms',	'sqft_living',	'sqft_lot',	'floors', 'waterfront',	'view',	'condition', 'sqft_above',	'sqft_basement', 'yr_built', 'yr_renovated']] 
```

```python
y=house_price[['price']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
```

```python
bedroom_groups = house_price.groupby('bedrooms')['price'].mean()

plt.figure(figsize=(8,5))
plt.bar(bedroom_groups.index, bedroom_groups.values, color='orange')
plt.title("Average Price by Number of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```
